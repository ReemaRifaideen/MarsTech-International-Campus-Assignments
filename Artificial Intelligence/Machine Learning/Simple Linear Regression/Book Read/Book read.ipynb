```python
#simple Linear regression Book-Read

import pandas as pd
#use pandas to reas csv format and create, read table
```

```python
#read csv file

dataset = pd.read_csv("book_read.csv")

#view dataset

dataset
```

```python
#split input data(x axis)

Feature = dataset [['HoursSpentReading']]
```

```python
#split output data(y axis)

Label = dataset [['BooksRead']]
```

```python
#split training dataset , testing dataset

from sklearn.model_selection import train_test_split
```

```python
#train dataset

x_train,x_test,y_train,y_test = train_test_split (Feature , Label , test_size=0.40 , random_state=0)
```

```python
#model creation, train and test as well

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#slop / m of LinearRegression

weight = regressor.coef_
weight
```

```python
#calculate bias / intercept of the model

bias = regressor.intercept_
bias 
```

```python
y_predict = regressor.predict(x_test)
y_predict
```

```python
#analyzing the prediction

from sklearn.metrics import r2_score

#test 'actual y' and 'predict y'
r2_score = r2_score (y_test, y_predict)
```

```python
r2_score
```
