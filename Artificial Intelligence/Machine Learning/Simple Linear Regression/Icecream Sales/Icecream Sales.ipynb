```python
#simple linear regression Icream-Sales

import pandas as pd
```

```python
 dataset =pd.read_csv("Icecream Sales - temperatures.csv") 
```

```python
x =dataset[["Temperature"]] #split input data(x axis)
```

```python
y =dataset[["Ice Cream Profits"]] #split output data(y axis)
```

```python
#split training data set testing dataset

from sklearn.model_selection import train_test_split 
```

```python
#train dataset

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.45,random_state=0)
```

```python
#model creation

from sklearn.linear_model import LinearRegression 

regressor=LinearRegression() 
regressor.fit(x_train,y_train)
```

```python
#slop of the linear regression

weight=regressor.coef_
weight
```

```python
#calculating the bias (intercept) of the model

bias=regressor.intercept_
bias
```

```python
#testing the model after trained

y_pred=regressor.predict(x_test)
y_pred
```

```python
#analyzing the prediction

from sklearn.metrics import r2_score 
#test 'actual y' and 'predict y'
r_score=r2_score(y_test,y_pred)   
```

```python
#view analyzed score

r_score 
```

```python
#import model

import pickle 

#save sav file and load sav file
sales="Icecream_Sales_temperatures.sav"
```

```python
#save model in .sav format

pickle.dump(regressor,open(sales,'wb'))
```
